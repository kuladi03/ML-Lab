```python
names =['Ram' , 'Sham' , 'Geeta']
marks = [87,50,98]
```

```python
fig , ax = plt.subplots(2,2)
ax[0,0].bar(names,marks)
ax[0,0].set_title('bar chart')
ax[0,1].scatter(names,marks)
ax[0,1].set_title('Scatter Chart')
ax[1,0].plot(names,marks)
ax[1,0].set_title('Plot')
ax[1,1].hist(marks)
ax[1,1].set_title('Histogram')
```

```python
x = np.arange(0.0 , 2.0 , 0.01)
y = 1 + np.sin(2 * np.pi * x)

fig , ((ax1,ax2),(ax3,ax4),(ax5,ax6)) = plt.subplots(3,2)
ax1.plot(x,y, color = "orange")
ax2.plot(x,y, color = "green")
ax3.plot(x,y, color = "blue")
ax4.plot(x,y, color = "magenta")
ax5.plot(x,y, color = "black")
ax6.plot(x,y, color = "red")
plt.rcParams['figure.figsize'] = [15,15]
```
